### Categorical Variables

```python
categorical_columns = df.loc[:,df.dtypes==np.object]
for c in categorical_columns:
    df[c] = df[c].astype('category')
    if df[c].isnull().any():
        df[c] = df[c].cat.add_categories(['MISSING'])
        df[c] = df[c].fillna('MISSING')

def boxplot(x, y, **kwargs):
    sns.boxplot(x=x, y=y)
    x=plt.xticks(rotation=90)
f = pd.melt(df, id_vars=['SalePrice'], value_vars=categorical_columns)
g = sns.FacetGrid(f, col="variable",  col_wrap=2, sharex=False, sharey=False, height=5)
g = g.map(boxplot, "value", "SalePrice")
```

### From numeric variables we picked elements where correlation between them and SalePrice is over 0.5. From categorical variables we picked based on boxplot above

```python
strong_numeric = df.corr()[df.corr()['SalePrice'] > 0.5]['SalePrice'].to_frame().T.columns[:-1]
categorical_variables = ['LotShape', 'LotConfig', 'Neighborhood', 'SaleType', 'BsmtFinType1']
```

```python
print(strong_numeric)
print(categorical_variables)
```

Below we define our X which consist of numerical and categorical variables we picked above

```python
X = pd.concat([df[strong_numeric], df[categorical_variables]], axis=1)
y = df.iloc[:, -1]
```

```python
X.head()
```

### Overall we have 15 columns in X and 1 in target (y)

### There are no missing values in columns we picked so no need for imputation( no need for part d but we will still use one hot encoding and scaling)

```python
X.isna().sum()
```

### Part B C defining the model

```python
param_space = {
    'regressor__n_neighbors': range(5, 50, 2)
}
```

```python
model = Pipeline(steps=[
    ('preprocessing', ColumnTransformer(transformers=[
        ('numeric', Pipeline(steps=[
            ('scaler', StandardScaler())
        ]), strong_numeric),
        ('categorical', Pipeline(steps=[
            ('ohe', OneHotEncoder())
        ]), categorical_variables)
    ])),
    ('regressor', KNeighborsRegressor(n_neighbors=15))
])
```

```python
gridsearch = GridSearchCV(model, param_space, cv=10)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y) 
```

```python
model.fit(X_train, y_train)
```

```python
print(model.score(X_test, y_test))
print(mean_squared_error(y_test, model.predict(X_test)))
```

### Above we can see mean squared error of KNeighborsRegressor 

### Part E: Now lets use different cross validations

```python
from sklearn.model_selection import KFold
kf = KFold(5, shuffle=True)
```

```python
accuracies = []
mse_errors = []
for train_idx, test_idx in kf.split(X, y):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y[train_idx], y[test_idx]
    
    model.fit(X_train, y_train)
    accuracies.append(model.score(X_test, y_test))
    mse_errors.append(mean_squared_error(y_test, model.predict(X_test)))
    
print(np.mean(accuracies))
print(np.std(accuracies))
print(np.mean(mse_errors))
```

### Part E: Here is the visualisation of our prediction model

```python
plt.scatter(model.predict(X_train), y_train, c = "blue",  label = "Training data")
plt.scatter(model.predict(X_test), y_test, c = "pink",  label = "Test data")
plt.title("Linear regression")
plt.xlabel("Predicted values")
plt.ylabel("Real values")
plt.legend(loc = "upper left")
plt.plot([100, 100], [100, 10], c = "red")
plt.show()
```
